**Stage1**

```python
f = open("group35-stage1.txt", 'r')
all_line = []
while True:
    line = f.readline()
    all_line.append(line)
    if not line: break
f.close()
```

```python
import re
data_class = pd.read_excel("./Annotation.xlsx")
data_class["Word"] = data_class["Word"].apply(lambda x : x.strip())
data_class["Word"] = data_class["Word"].apply(lambda x : re.sub("[^a-zA-Z( )]","",x.lower()))
data_class
```

```python
res = {}
i = 0
for l in all_line:
    tmp = re.sub("[^a-zA-Z( )]","",l).lower().split(" ")
    j = 0
    for w in tmp:
        if w in np.array(data_class["Word"]):
            res[((i,j),(i,j))] = data_class["Annotation"][np.array(data_class["Word"]) == w].values[0] #+ " " + w
        j += 1

    for k in range(1,6):
        j = 0
        for ab in range(len(tmp)-k):
            w = " ".join(tmp[ab:(ab+(k+1))])
            if w in np.array(data_class["Word"]):
                res[((i,j),(i,j+k))] = data_class["Annotation"][np.array(data_class["Word"]) == w].values[0] #+ " " + w
            j += 1
    i += 1

```

```python
my_anno = pd.DataFrame({"key" : list(res.keys()), "NER" :list(res.values())})
my_anno
```

```python
#줄바꿈 경우의 수 추가
newline = pd.DataFrame({"key": [((263, 8), (264, 0)), ((262, 7), (263,0))],
                        "NER": ["GPE", "GPE"]})
my_anno = pd.concat([my_anno,newline])
```

```python
my_anno = my_anno.reset_index(drop = True)
my_anno
```

```python
f = open("group35-stage1.txt.provided-anno.txt", 'r')
anno_data = []
while True:
    line = f.readline()
    anno_data.append(line)
    if not line: break
f.close()

anno_data[1:3]
```

```python
f = open("group35-stage1.txt", 'r')
org_data = []
while True:
    line = f.readline()
    org_data.append(line)
    if not line: break
f.close()
ith_len = [len(org_data[i].split(" ")) for i in range(len(org_data))][:-1]
ith_len[:5]
```

```python
res = {} #i는 줄, j는 열의 갯수
for i in range(len(org_data)):
    tmp = org_data[i].split(" ")
    for j in range(len(tmp)):
        res[(i,j)] = tmp[j]
res
```

```python
my_anno
```

```python
import re
w_all = []
num = my_anno.key
w_all = []
for k in range(len(num)):
    print(k)
    split_annotate_num = num[k]
    i0 = int(split_annotate_num[0][0])
    j0 = int(split_annotate_num[0][1])
    i1 = int(split_annotate_num[1][0])
    j1 = int(split_annotate_num[1][1])

    if (i0 == i1) and (j0 == j1):
        w = res[(i0,j0)]
    elif i0 == i1:
        w = ""
        for j in range(j0,j1+1):
            w += " "+res[(i0,j)]
    else:
        w = ""
        for i in range(i0,i1+1):
            if i == i0:
                for j in range(j0,ith_len[i]):
                    w += " "+res[(i,j)]
            elif i == i1:
                for j in range(0,j1+1):
                    w += " "+res[(i,j)]
            else :
                for j in range(0,ith_len[i]):
                    w += " "+res[(i,j)]

    w_all.append(re.sub("[^a-zA-Z( )]","",w).lower().strip())
w_all


```

```python
annotated_step2_my = pd.DataFrame({"word":w_all,"ano":my_anno.NER,"num" : num})
annotated_step2_my["word"] = annotated_step2_my["word"].apply(lambda x : re.sub("[^a-zA-Z( )]","",x.lower()))
annotated_step2_my = annotated_step2_my.drop_duplicates()
annotated_step2_my["num"] = annotated_step2_my["num"].apply(lambda x : str(x))
annotated_step2_my
```

```python
# 교수님꺼
num = list(map(lambda x : x.split("-")[0].strip() if len(x.split("-")) == 2 else "",anno_data))[:-1]
ano = list(map(lambda x : x.split("-")[1].strip() if len(x.split("-")) == 2 else "",anno_data))[:-1]

w_all = []
for k in range(len(num)):
    split_annotate_num = re.sub("[^0-9,]","",num[k]).split(",")
    i0 = int(split_annotate_num[0])
    j0 = int(split_annotate_num[1])
    i1 = int(split_annotate_num[2])
    j1 = int(split_annotate_num[3])

    if (i0 == i1) and (j0 == j1):
        w = res[(i0,j0)]
    elif i0 == i1:
        w = ""
        for j in range(j0,j1+1):
            w += " "+res[(i0,j)]
    else:
        w = ""
        for i in range(i0,i1+1):
            if i == i0:
                for j in range(j0,ith_len[i]):
                    w += " "+res[(i,j)]
            elif i == i1:
                for j in range(0,j1+1):
                    w += " "+res[(i,j)]
            else :
                for j in range(0,ith_len[i]):
                    w += " "+res[(i,j)]

    w_all.append(re.sub("[^a-zA-Z( )]","",w).lower().strip())
w_all

```

```python
annotated_step2_professor = pd.DataFrame({"word_p":w_all,"ano_p":ano,"num" : num})
annotated_step2_professor["word_p"] = annotated_step2_professor["word_p"].apply(lambda x : re.sub("[^a-zA-Z( )]","",x.lower()))
annotated_step2_professor = annotated_step2_professor.drop_duplicates()
annotated_step2_professor
```

```python
outer = pd.merge(annotated_step2_my,annotated_step2_professor, on = "num", how = "outer")
outer.to_excel("outer.xlsx")
outer
```

**(a)**

```python
TP = len(outer[pd.notnull(outer['ano']) & (outer['ano'] == outer['ano_p'])])

FP = len(outer[pd.notnull(outer['ano']) & (outer['ano'] != outer['ano_p'])])

FN = len(outer[pd.isnull(outer['ano'])])

#
TN = 0

TP + FP + FN + TN
```

**(b)**

```python
P = TP / (TP + FP)
R = TP / (TP + FN)

F1 = 2 *  (P * R) / (P + R)
F1
```
